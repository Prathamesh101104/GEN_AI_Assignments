```python
# Function to generate images using Stability AI's Stable Diffusion model
def generate_image_sd(prompt, size=1024):
    
    # API endpoint for Stability AI image generation
    url = "https://api.stability.ai/v2beta/stable-image/generate/core"

     # HTTP request headers containing authentication information
    headers = {
        "Authorization": f"Bearer {STABILITY_API_KEY}",
        "Accept": "image/*"
    }

     # Payload parameters sent to the API
     # 'files' is used because the API expects multipart form data
    files = {
        "prompt": (None, prompt),  
        "width": (None, str(size)),
        "height": (None, str(size)),
        "output_format": (None, "png")
    }
    
     # Sending POST request to Stability AI server
    response = requests.post(url, headers=headers, files=files)
    
     # Check if request failed
    if response.status_code != 200:
        raise Exception("API Error: " + response.text)
    
    # Convert API response into an image object
    image_bytes = response.content
    
    # Return generated image as a list
    img = Image.open(BytesIO(image_bytes)).convert("RGBA")

    return [img]
```

```python
#Prompt Definition
prompt = (
    "A high-resolution cinematic photograph of a red fox in a mossy forest, "
    "golden hour lighting, shallow depth of field, Nikon 85mm style."
)
```

```python
#Generate Image
images = generate_image_sd(prompt)
```

```python
#Save Generated Image
for i, img in enumerate(images):
    img.save(f"sd_out_{i+1}.png")
    print("Saved sd_out_", i+1)
```
